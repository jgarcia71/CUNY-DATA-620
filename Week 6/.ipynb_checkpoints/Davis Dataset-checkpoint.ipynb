# Davis Dataset

A brief assignment by Joy Payton

These data were collected by Davis et al. in the 1930s.
They represent observed attendance at 14 social events by 18 Southern women.
The graph is bipartite (events, women).

The question we are asked:  "What can you infer about the relationships between (1) the women, and (2) the social events?"

## Set up data

```python
import networkx as nx
import networkx.algorithms.bipartite as bipartite

G = nx.davis_southern_women_graph()
```

Let's define our sets!

```python
women, events = bipartite.sets(G)
print "\nWomen:\n" + str(list(women))
print "\nEvents:\n" + str(list(events))
```

## Visualization

Let's take a quick peek at this graph!  We'll draw the women as smaller red nodes (with size relative to their degree centrality) and the social events as larger blue nodes (again, with differential sizes).

```python
import matplotlib.pyplot as plt
%matplotlib inline

pos=nx.spring_layout(G) # positions for all nodes

# calculate degree centrality
womenDegree = nx.degree(G, women) 
eventsDegree = nx.degree(G, events)

plt.figure(1,figsize=(15,10)) 

# nodes
nx.draw_networkx_nodes(G,pos,
                       nodelist=women,
                       node_color='r',
                       node_size=[v * 50 for v in womenDegree.values()],
                   alpha=0.8)
nx.draw_networkx_nodes(G,pos,
                       nodelist=events,
                       node_color='b',
                       node_size=[v * 100 for v in eventsDegree.values()],
                   alpha=0.8)

# edges
nx.draw_networkx_edges(G,pos,width=1.0,alpha=0.5)
```

## Women

Let's concentrate first on women!

### Basic Centrality Measures

Well, we certainly see that some women have greater connectedness than others.  Let's check out the relative centrality, which we already calculated for plotting purposes.

```python
import pandas as pd

womenDegreeDF = pd.DataFrame(womenDegree.items())
womenDegreeDF.columns = [
  'woman',
  'degree'
]
womenDegreeDF.sort_values(by='degree', ascending = False)
```

It seems like Theresa Anderson, Evelyn Jefferson, Nora Fayette, Sylvia Avondale, Laura Mandeville, and Brenda Rogers are the social butterflies of the group!

### Going a bit deeper

Do our "social butterflies", who frequent social events, have more acquaintanceships and interact with more women?  Or are some of them "clique-ish" and always see the same people over and over?  Let's look at acquaintanceships!

We can do this by projecting our bipartite graph onto women nodes.  We'll first check it out graphically and then numerically.

```python
W = bipartite.projected_graph(G, women)
plt.figure(2,figsize=(15,10)) 
nx.draw_shell(W, with_labels = True)
```

Wow, that's lovely, but doesn't tell us much.  Let's crunch the numbers!

```python
womenAcquaintanceship = pd.DataFrame({'woman': [w for w in women], \
                                'acquaintance_count': [W.degree(w) for w in women]})
womenAcquaintanceship.sort_values(by='acquaintance_count', ascending = False)
```

We have seven women who have socialized with every other woman in our data set:  Theresa Anderson, Helen Lloyd, Evelyn Jefferson, Nora Fayette, Sylvia Avondale, Verne Sanderson, and Ruth DeSand.  

### Combine Data 

```python
womenSocialActivity = womenDegreeDF.merge(womenAcquaintanceship)
womenSocialActivity.sort_values(by=['degree', 'acquaintance_count'], ascending = False)
```

### Ego Networks

```python
dorothy = nx.Graph(nx.ego_graph(G, 'Dorothy Murchison', radius = 2))
plt.figure(2,figsize=(15,10)) 
nx.draw(dorothy, with_labels = True)
```

Dorothy attended two events, E8 and E9, which had great coverage between the two.  Let's compare her to Charlotte McDowd, who attended twice as many events, but has a lower number of acquaintances:

```python
charlotte = nx.Graph(nx.ego_graph(G, 'Charlotte McDowd', radius = 2))
plt.figure(3,figsize=(15,10)) 
nx.draw(charlotte, with_labels = True)
```

Charlotte's ego network is much more tightly interconnected -- she socializes at events which tend to have some of the same people.  It looks like Brenda Rogers, Theresa Anderson, and Evelyn Jefferson are often found the same places as Charlotte.  We've arrived at this "clique" definition by a bit of human detective work... will networkx's algorithmic approach  yield similar findings?

```python
list(nx.find_cliques(W))
```

The "cliques" are quite large and captured more people than we expected (each 'clique' constitutes the majority of the women!).  Can we find clusters that might be more useful?  Clusters will separate women into distinct groups based on a threshold value we can tweak.

### Clustering

I'm taking code here from Chapter 4 of _Social Network Analysis for Startups_ by Maksim Tsvetovat and Alexander Kouznetsov (see pp 85-86).

```python
import numpy
from scipy.cluster import hierarchy
from scipy.spatial import distance
from collections import defaultdict

def create_hc(G, t=1.0):  # default threshold is 1 -- we'll play with this
    labels = G.nodes()
    path_length = nx.nx.all_pairs_shortest_path_length(G)
    distances = numpy.zeros((len(G), len(G)))
    i = 0
    for u,p in path_length.items():
        j=0
        for v,d in p.items():
            distances[i][j] = d
            distances[j][i] = d
            if i==j: 
                distances [i][j] = 0
            j+=1
        i+=1    
    Y = distance.squareform(distances)
    Z = hierarchy.complete(Y)
    membership = list(hierarchy.fcluster(Z, t=t))
    partition = defaultdict(list)
    for n,p in zip(list(range(len(G))), membership):
        partition[p].append(labels[n])
    return list(partition.values())
```

```python
create_hc(W, 1)
```

Based on a threshold of 1, we see two clusters of women who appear together predictably.  What happens if we lower the threshold?

```python
create_hc(W, 0)
```

Hm. It's sort of interesting to see what might be friendships between pairs of women, like Eleanor Nye and Brenda Rogers.  Note that we have some loners, like Frances Anderson, and we have one large cluster which includes my good friend Dorothy!

## Social Events

Similar to what we did with women, let's see what we can figure out about events!

### Centrality

```python
eventsDegreeDF = pd.DataFrame(eventsDegree.items())
eventsDegreeDF.columns = [
  'event',
  'degree'
]
eventsDegreeDF.sort_values(by='degree', ascending = False)
```

```python
### Visualization
```

It seems that social events E5-E9 are the most popular.  Can we visualize the links between events in terms of "have attendees in common"?

```python
E = bipartite.projected_graph(G, events)
plt.figure(4,figsize=(15,10)) 
nx.draw_shell(E, with_labels = True)
```

Similar to our visualization of women alone, this doesn't tell us that much -- it's a bit confusing!  We'll analyze this in a different way:

```python
eventOverlap = pd.DataFrame({'event': [e for e in events], \
                                'shares_attendees_count': [E.degree(e) for e in events]})
eventOverlap.sort_values(by='shares_attendees_count', ascending = False)
```

There are really only two circumstances at play here: 

* Some events have attendees in common with all of the other events (total of 13)
* Some events only have attendees in common with 8 other events (so are likely more lightly attended!)

### Combine Data

```python
eventConnectedness = eventsDegreeDF.merge(eventOverlap)
eventConnectedness.sort_values(by=['degree', 'shares_attendees_count'], ascending = False)
```

As expected, events that are less well connected to other events by common attendees have lower attendance.  The pattern here is a bit linear than the data we saw with women.  As the number of attendees grows, the number of events that a given event is linked to by one or more common attendess also grows.  Nothing terribly surprising here!  It doesn't seem useful to do things like clique or cluster analysis.
